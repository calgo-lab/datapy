# Java vs. Python

### Compile vs. interpret

Java:
- A Java program is complied -- it can be executed only after the *entire* program is successfully compiled

Python:
- Conceptually, a Python program is interpreted -- it is "directly" executed in a line-by-line manner, and program halts when an error occurs
- Fun fact -- you can execute a Python program even when it has (e.g. syntax) errors!
- The technical details are more complicated (at least in non-interactive mode):
  - The program is compiled (yes, indeed) first into bytecode
  - Then a virtual machine executes this bytecode
  - In a line-by-line manner, and halts when an error occurs

### One main version vs. multiple virtual environments

Java:
- One (primary) version is installed in the computer
- All softwares including IDEs (Eclipse, IntelliJ, etc.) use this verion

Python:
- In one virtual environment, there is one specific version of Python, together with specific versions of libraries
- Multiple virtual environments are allowed

Java:

Python:

# Virtual environments

- For different projects you will need different dependencies (or versions thereof)
    
- It is good practice to encapsulate your dependencies in a virtual environment


When you see a venv's name in brackets at the beginning of a line in the shell:

```
(base) C:\Users\Butter Ball>
```
You know that that specific venv (called "based" in the example above) has been activated.

Create a new venv, activate and deactivate it:
```
python3 -m venv venv
source venv/bin/activate
deactivate
```

Or with Anaconda:

```
conda create -n venv python=3.9
conda activate venv
conda deactivate
```

### How to run Python code...

- ... not interactively:
  - Method 1: execute entire program via shell command ```$ python [somefilename].py```

- ... interactively:
  - Method 2: Python interpreter ```$ python```
  - Method 3: IPython interpreter ```$ ipython```
  - Method 4: Jupyter Notebook (IPython interpreter served in the web browser)

We will discuss pros and cons after looking into each of these options

## Method 1: execute entire program via shell command

```python
python some_file.py
hello world

```

## Method 2: Python interpreter

Type ``python`` at the command prompt:
```
$ python
Python 3.5.1 |Continuum Analytics, Inc.| (default, Dec  7 2015, 11:24:55)
Type "help", "copyright", "credits" or "license" for more information.
>>>
```
You can type and execute code snippets:
``` python
>>> 1 + 1
2
>>> x = 5
>>> x * 3
15
```

You can run python code files:
```python
exec(open('some_file.py').read())
```

## Method 3: IPython interpreter

The Python shell lacks many useful features of a modern shell:

- tab-completion
- history (per default)
- help/documentation functionality
- plotting

IPython adds all that.


To start IPython, type ``ipython`` at the command prompt:

```
$ ipython
Python 3.5.1 |Continuum Analytics, Inc.| (default, Dec  7 2015, 11:24:55) 
Type "copyright", "credits" or "license" for more information.

IPython 4.0.0 -- An enhanced Interactive Python.
?         -> Introduction and overview of IPython's features.
%quickref -> Quick reference.
help      -> Python's own help system.
object?   -> Details about 'object', use 'object??' for extra details.

In [1]: 
```


Notice the visual difference:
- Python uses ``>>>`` 
- IPython uses numbered commands (e.g. ``In [1]:``)

IPython has the same functionalities as a standard Python shell:
``` ipython
In [1]: 1 + 1
Out[1]: 2

In [2]: x = 5

In [3]: x * 3
Out[3]: 15
```

... and **many** more

You can run python code files:
```python
%run some_file.py
```

## Some nice IPython features

## Method 4: Jupyter Notebook

- made for Julia, python and R (hence **Ju**-**py**-te**R**)
- has all functionality of the IPython shell
- runs in browser
- simpler plotting
- sharing of analyses simpler (even dashboards)
- *extremely* popular (also in companies)


## The Jupyter Notebook

Adds many useful features:
 - remote execution
 - inline plotting
 - interactive plots
 
But has some drawbacks:
 - hidden state
 - version control is difficult
 - testing is difficult

### Pros and cons of execution methods

### Method 1: execute entire program via shell command

**Pros**
- version control
- clean reproducible execution (e.g. no hidden state)
- precise timing

**Cons**
- not interactive
- debugging more difficult


### Method 2: Python interpreter

**Pros**
- interactive development

**Cons**
- very inconvenient

I only use it if there is no IPython installed (and there's no internet).


### Method 3: IPython interpreter

**Pros**
- interactive development / debugging
- convenience
 - documentation
 - debugging
 - tab completion
 - history

**Cons**
- hidden state



### Method 4: Jupyter Notebook

**Pros**
- all of IPython's advantages
- inline plotting
- code and markdown cells allow better documentation

**Cons**
- too convenient (for people too lazy to organize code)
- hidden state
- version control is more difficult
