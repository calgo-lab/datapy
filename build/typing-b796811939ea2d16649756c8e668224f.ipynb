## 1. Introduction to Type Hints

Python is a **dynamically typed** language, meaning you don't need to declare variable types. However, **type hints** (also called **type annotations**) allow you to:

✅ Improve code readability  
✅ Enable better IDE autocomplete and error detection  
✅ Facilitate static type checking (using tools like `mypy`)  
✅ Serve as documentation  

**Important**: Type hints are **optional** and don't affect runtime behavior. Python remains dynamically typed.

````python
# Without type hints
def greet(name):
    return f"Hello, {name}!"

# With type hints
def greet_typed(name: str) -> str:
    return f"Hello, {name}!"

print(greet("Alice"))
print(greet_typed("Bob"))

# Type hints don't prevent this from running (but mypy would flag it)
print(greet_typed(123))  # Works at runtime!
````

## 2. Basic Type Annotations

### Built-in Types

Python's built-in types can be used directly for annotations:

| Type | Description | Example |
|------|-------------|----------|
| `int` | Integer | `age: int = 25` |
| `float` | Floating point | `price: float = 19.99` |
| `str` | String | `name: str = "Alice"` |
| `bool` | Boolean | `is_active: bool = True` |
| `bytes` | Byte string | `data: bytes = b"hello"` |
| `None` | None type | `result: None = None` |

````python
# Variable annotations
name: str = "Alice"
age: int = 30
height: float = 5.6
is_student: bool = False

print(f"{name} is {age} years old, {height} feet tall")
print(f"Student status: {is_student}")

# You can also annotate without assignment
country: str
country = "USA"  # Assigned later
print(f"Country: {country}")
````

## 3. Collection Types

For collections, we use the `typing` module to specify the types of elements they contain.

### Python 3.9+ (Simplified Syntax)

````python
# Python 3.9+ allows using built-in collection types directly
from typing import Dict, List, Set, Tuple  # Still works, but not necessary in 3.9+

# Lists
numbers: list[int] = [1, 2, 3, 4, 5]
names: list[str] = ["Alice", "Bob", "Charlie"]

# Dictionaries
scores: dict[str, int] = {"Alice": 95, "Bob": 87}
config: dict[str, str] = {"host": "localhost", "port": "8080"}

# Sets
unique_ids: set[int] = {1, 2, 3, 4, 5}

# Tuples (fixed size)
coordinates: tuple[float, float] = (10.5, 20.3)
rgb_color: tuple[int, int, int] = (255, 128, 0)

# Tuples (variable size)
variable_tuple: tuple[int, ...] = (1, 2, 3, 4, 5)

print(f"Numbers: {numbers}")
print(f"Scores: {scores}")
print(f"Coordinates: {coordinates}")
````

### Python 3.5-3.8 (Using typing module)

````python
from typing import List, Dict, Set, Tuple

# For Python < 3.9, use capital letter versions from typing
old_style_list: List[int] = [1, 2, 3]
old_style_dict: Dict[str, float] = {"pi": 3.14, "e": 2.71}
old_style_set: Set[str] = {"apple", "banana", "cherry"}
old_style_tuple: Tuple[str, int, bool] = ("Alice", 30, True)

print(f"Old style list: {old_style_list}")
print(f"Old style dict: {old_style_dict}")
````

## 4. Optional and Union Types

### Optional Type
`Optional[X]` means the value can be of type `X` or `None`. It's equivalent to `Union[X, None]`.

### Union Type
`Union[X, Y]` means the value can be of type `X` **or** `Y`.

### Python 3.10+ Syntax
Python 3.10 introduced the `|` operator for union types.

````python
from typing import Optional, Union

# Optional: can be str or None
def get_middle_name(full_name: str) -> Optional[str]:
    parts = full_name.split()
    if len(parts) == 3:
        return parts[1]
    return None

print(get_middle_name("John Fitzgerald Kennedy"))  # "Fitzgerald"
print(get_middle_name("John Kennedy"))             # None

# Union: can be int or str
def process_id(user_id: Union[int, str]) -> str:
    return f"Processing ID: {user_id}"

print(process_id(12345))
print(process_id("ABC-123"))

# Python 3.10+ syntax (using | operator)
def modern_union(value: int | str | None) -> str:
    return f"Value: {value}"

print(modern_union(42))
print(modern_union("hello"))
print(modern_union(None))
````

## 5. Type Aliases

Type aliases make complex type hints more readable by assigning them to variables.

````python
from typing import Union, List, Dict

# Create type aliases
UserId = int
Username = str
UserData = Dict[str, Union[str, int]]
Matrix = List[List[float]]

def get_user(user_id: UserId) -> UserData:
    return {
        "id": user_id,
        "name": "Alice",
        "age": 30
    }

def multiply_matrix(a: Matrix, b: Matrix) -> Matrix:
    # Simplified matrix multiplication
    return [[1.0, 2.0], [3.0, 4.0]]

user = get_user(12345)
print(f"User: {user}")

matrix_a: Matrix = [[1.0, 2.0], [3.0, 4.0]]
matrix_b: Matrix = [[5.0, 6.0], [7.0, 8.0]]
result = multiply_matrix(matrix_a, matrix_b)
print(f"Matrix result: {result}")
````

## 6. Function Annotations

Function type hints specify:
- **Parameter types**: Types of function arguments
- **Return type**: Type of the returned value (using `->` syntax)

### Special Return Types
- `-> None`: Function doesn't return anything
- `-> NoReturn`: Function never returns (e.g., raises exception or infinite loop)

````python
from typing import NoReturn

# Basic function with type hints
def add(a: int, b: int) -> int:
    return a + b

# Function with multiple parameters
def format_person(name: str, age: int, city: str = "Unknown") -> str:
    return f"{name}, {age}, from {city}"

# Function that returns nothing
def log_message(message: str) -> None:
    print(f"LOG: {message}")

# Function that never returns
def raise_error(message: str) -> NoReturn:
    raise ValueError(message)

# Function with complex return type
def get_stats(numbers: list[int]) -> dict[str, float]:
    return {
        "mean": sum(numbers) / len(numbers),
        "min": float(min(numbers)),
        "max": float(max(numbers))
    }

print(add(5, 3))
print(format_person("Alice", 30, "NYC"))
log_message("System started")
print(get_stats([1, 2, 3, 4, 5]))

# Uncomment to see error:
# raise_error("Something went wrong!")
````

## 7. Classes and Type Hints

Type hints work with class attributes, methods, and instance variables.

````python
from typing import ClassVar, Optional

class Person:
    # Class variable
    species: ClassVar[str] = "Homo sapiens"
    
    # Instance variables with type hints
    name: str
    age: int
    email: Optional[str]
    
    def __init__(self, name: str, age: int, email: Optional[str] = None) -> None:
        self.name = name
        self.age = age
        self.email = email
    
    def greet(self) -> str:
        return f"Hello, I'm {self.name}, {self.age} years old."
    
    def has_email(self) -> bool:
        return self.email is not None
    
    @classmethod
    def from_birth_year(cls, name: str, birth_year: int) -> 'Person':
        """Alternative constructor"""
        age = 2026 - birth_year
        return cls(name, age)

# Create instances
alice = Person("Alice", 30, "alice@example.com")
bob = Person.from_birth_year("Bob", 1990)

print(alice.greet())
print(f"Alice has email: {alice.has_email()}")
print(bob.greet())
print(f"Species: {Person.species}")
````

## 8. Advanced Topics

### Callable Types
For functions as arguments or return values.

````python
from typing import Callable

# Function that takes a function as argument
def apply_operation(x: int, y: int, operation: Callable[[int, int], int]) -> int:
    return operation(x, y)

def add(a: int, b: int) -> int:
    return a + b

def multiply(a: int, b: int) -> int:
    return a * b

print(apply_operation(5, 3, add))       # 8
print(apply_operation(5, 3, multiply))  # 15

# Function that returns a function
def make_multiplier(n: int) -> Callable[[int], int]:
    def multiplier(x: int) -> int:
        return x * n
    return multiplier

times_three = make_multiplier(3)
print(times_three(5))  # 15
````

### Generic Types (TypeVar)
For writing generic functions and classes.

````python
from typing import TypeVar, List

T = TypeVar('T')  # Generic type variable

def get_first_element(items: List[T]) -> T:
    """Returns the first element of a list."""
    return items[0]

# Works with different types
numbers = [1, 2, 3]
names = ["Alice", "Bob", "Charlie"]

first_num: int = get_first_element(numbers)      # Type checker knows it's int
first_name: str = get_first_element(names)       # Type checker knows it's str

print(f"First number: {first_num}")
print(f"First name: {first_name}")

# Generic class
class Stack(List[T]):
    def push(self, item: T) -> None:
        self.append(item)
    
    def pop_item(self) -> T:
        return self.pop()

int_stack: Stack[int] = Stack()
int_stack.push(1)
int_stack.push(2)
print(f"Popped: {int_stack.pop_item()}")
````

### Literal Types
For specifying exact values allowed.

````python
from typing import Literal

# Only specific string values allowed
def set_mode(mode: Literal["read", "write", "append"]) -> str:
    return f"Mode set to: {mode}"

print(set_mode("read"))   # OK
print(set_mode("write"))  # OK
# set_mode("delete")      # Type checker would flag this as error

# Literal with numbers
def roll_dice() -> Literal[1, 2, 3, 4, 5, 6]:
    import random
    return random.choice([1, 2, 3, 4, 5, 6])  # type: ignore

print(f"Dice roll: {roll_dice()}")
````

### Protocol (Structural Subtyping)
Define interfaces based on structure rather than inheritance.

````python
from typing import Protocol

class Drawable(Protocol):
    """Any object with a draw() method"""
    def draw(self) -> str:
        ...

class Circle:
    def draw(self) -> str:
        return "Drawing a circle"

class Square:
    def draw(self) -> str:
        return "Drawing a square"

def render(shape: Drawable) -> None:
    """Works with any object that has draw() method"""
    print(shape.draw())

# Both work even though they don't inherit from Drawable
render(Circle())
render(Square())
````

## 9. Type Checking with mypy

**mypy** is a static type checker for Python. It analyzes your code without running it.

### Installation
```bash
pip install mypy
```

### Usage
```bash
mypy your_script.py
```

### Example: Finding Type Errors

````python
# This code runs fine, but mypy would catch the type error

def calculate_area(radius: float) -> float:
    return 3.14159 * radius ** 2

# Correct usage
area1 = calculate_area(5.0)
print(f"Area 1: {area1}")

# Type error: passing string instead of float
# This runs in Python but mypy would flag it:
area2 = calculate_area("5")  # type: ignore
print(f"Area 2: {area2}")  # Would crash at runtime

# Running: mypy this_file.py
# Output: error: Argument 1 to "calculate_area" has incompatible type "str"; expected "float"
````

## Best Practices

✅ **Start gradually** — Add type hints to new code and critical functions first  
✅ **Use type checkers** — Integrate `mypy` into your CI/CD pipeline  
✅ **Be consistent** — If you start typing, type everything in that module  
✅ **Use type aliases** — Make complex types readable  
✅ **Document with types** — Type hints serve as inline documentation  
✅ **Use `Optional`** — Be explicit about values that can be `None`  
✅ **Leverage IDE support** — Modern IDEs use type hints for autocomplete  

❌ **Don't ignore errors** — If mypy flags something, investigate  
❌ **Don't over-complicate** — Use `Any` when types are truly dynamic  
❌ **Don't rely on runtime** — Type hints don't enforce types at runtime

## Summary

| Concept | Syntax | Purpose |
|---------|--------|----------|
| Basic types | `x: int`, `y: str` | Annotate variables |
| Collections | `list[int]`, `dict[str, int]` | Typed containers |
| Optional | `Optional[str]` or `str \| None` | Can be None |
| Union | `Union[int, str]` or `int \| str` | Multiple types |
| Callable | `Callable[[int, int], int]` | Function types |
| TypeVar | `T = TypeVar('T')` | Generic types |
| Literal | `Literal["a", "b"]` | Specific values |
| Protocol | `class P(Protocol)` | Structural typing |
