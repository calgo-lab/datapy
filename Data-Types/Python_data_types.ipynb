## Python Variables are passed-by-reference
In contrast, Java variables are passed-by-value

````python
x = 3.1415   # define a container in memory with value 3.1415 and assign variable/pointer x to it
y = x        # assign a second variable y that points to the same location as x
print(x)
print(y)
print(id(x)) # address of the container x is pointing at
print(id(y))
````

````python
x = 2.71     # define a new container with value 2.71 and reassign pointer x to the new location
print(x)
print(y)
print(id(x))
print(id(y))
````


**Note**: Java would have modified the same container *x* changing the value 3.1415 to 2.71.

In contrast, Python creates 2 containers (one with value 3.1415, one with value 2.71) and just reassigns the pointer *x*.

````python
# Bonus:
x += 3       # this also creates another container with the value 2.71 + 3 = 5.71
print(x)
print(id(x))
````

## Assignment Operations

| $~ $     | $~ $           | $~ $                                 |
|--------------|-----------------|---------------------------------------------|
|``a += b``| ``a -= b``|``a *= b``| ``a /= b``|
|``a //= b``| ``a %= b``|``a **= b``|``a &= b``|
|<code>a &#124;= b</code>| ``a ^= b``|``a <<= b``| ``a >>= b``|

````python
a = 2
print(id(a))

# 'Normal' Assignment
a = a + 2  # equivalent to a = a + 2, but more efficient (it only evaluates 'a' once)

print(a)
print(id(a)) # new address
````

````python
a = 2
print(id(a))

# Augmented Assignment
a += 2  # equivalent to a = a + 2, but more efficient (it only evaluates 'a' once)

# However: still new container created (integers are immutable)

print(a)
print(id(a)) # new address
````

### Pointers to Lists: Shallow vs Deep Copies

#### Scenario 1

````python
# What's happening here?
a = [1, 2, 3, 5] 
print(a)
b = a       
print(b)    
a = a + [8] 
            
print(a)    
print(b)
````

#### Scenario 1

````python
# What's happening here?
a = [1, 2, 3, 5] # Define a list container and point at it through pointer a
print(a)
b = a       # A new pointer b pointing to the same container as pointer a
print(b)    # Pointer b points to the same container as a
a = a + [8] # From right to left: access container from a, extend it with '8', store somewhere else in memory 
            # and reassign pointer a to the new location 
print(a)    # Pointer a points to the new extended container
print(b)    # Pointer b points to the original container
````

*Note*: Adding / concatenating lists together via `+` creates a new list.

#### Scenario 1.1

````python
# What's happening here?
a = [1, 2, 3, 5] 
print(a)
b = a       
print(b)    
a += [8]    

print(a)    
print(b)    
````

#### Scenario 1.1

````python
# What's happening here?
a = [1, 2, 3, 5] # Define a list container and point at it through pointer a
print(a)
b = a       # A new pointer b pointing to the same container as pointer a
print(b)    # Pointer b points to the same container as a
a += [8]    # Augmented Assignment: the original container is modified in-place without creating a new one

print(a)    # Pointer a points to the original (now extended) container
print(b)    # Pointer b points to the original (now extended) container
````

#### Scenario 2

````python
# What's happening here?
a = [1, 2, 3, 5] 
print(a)
b = a       
print(b)    
a.append(8)          
print(a)    
print(b)
````

#### Scenario 2

````python
# What's happening here?
a = [1, 2, 3, 5] # Define a list container and point at it through pointer a
print(a)
b = a       # A new pointer b pointing to the same container as pointer a
print(b)    # Pointer b points to the same container as a
a.append(8) # Access container from a and append value '8' to the same container / in the same location
print(a)    # Pointer a points to the same modified container
print(b)    # Pointer b also points to the same modified container
````

*Note*: Applying list-specific methods on list objects (`append`, `pop`, `insert`...) does NOT create a new container in memory; the same old container is being modified.

#### Scenario 3

````python
# What's happening here?
a = [1, 2, 3, 5] 
print(a)
b = a       
print(b)    
b[0] = 7        
print(a)    
print(b)
````

#### Scenario 3

````python
# What's happening here?
a = [1, 2, 3, 5] # Define a list container and point at it through pointer a
print(a)
b = a     # A new pointer b pointing to the same container as pointer a
print(b)  # Pointer b points to the same container as a
b[0] = 7  # Modify first item in the container via pointer b
print(a)  # Pointer a points to the same modified container
print(b)  # Pointer b also points to the same modified container
````

*Note*: Accessing list items through `[ ]` operates directly on the same container; no copies created.

#### Scenario 4

````python
# What's happening here?
a = [1, 2, 3, 5] 
print(a)
b = a[:]    
print(b)    
b[0] = 7        
print(a)    
print(b)
````

#### Scenario 4

````python
# What's happening here?
a = [1, 2, 3, 5] # Define a list container and point at it through pointer a
print(a)
b = a[:]  # !!! Create a copy of the container, store somewhere else in memory and point at it via pointer b
print(b)  # Pointer b points to a DIFFERENT container that just has the same list values as a's container
b[0] = 7  # Modify first item in the SECOND container via pointer b
print(a)  # Pointer a still points to the first container 
print(b)  # Pointer b still points to the second container
````

*Note*: Blank slicing a list - `[:]` - creates a copy of that list.

## Sets
- Unordered collections of unique items
- Support set operations
- Defined by ``{1, 2, 3}`` 

````python
primes = {2, 3, 5, 7}
odds = {1, 3, 5, 7, 9}
````

### Union
items appearing in either set

````python
primes = {2, 3, 5, 7}
odds = {1, 3, 5, 7, 9}
primes | odds      # with an operator
primes.union(odds) # equivalently with a method
````

### Intersection
items appearing in both sets

````python
primes = {2, 3, 5, 7}
odds = {1, 3, 5, 7, 9}
primes & odds             # with an operator
primes.intersection(odds) # equivalently with a method
````

### Difference
items appearing in one but not other set

````python
primes = {2, 3, 5, 7}
odds = {1, 3, 5, 7, 9}
primes - odds           # with an operator
primes.difference(odds) # equivalently with a method
````

### Symmetric difference
items appearing in only one set

````python
primes = {2, 3, 5, 7}
odds = {1, 3, 5, 7, 9}
primes ^ odds                     # with an operator
primes.symmetric_difference(odds) # equivalently with a method
````
